```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
```

```python
np.random.randn(10)
```

```python
x = np.linspace(0,1,10)
y = 2*x + 0.1*np.random.randn(10)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,1,10)
y = 2*x + 0.1*np.random.randn(10)

fig=plt.figure()
ax = fig.add_subplot(111)

ax.plot(x,y)
plt.show()
```

```python
x_1 = np.linspace(0,1,10)
y_1 = 2*x_1 + 0.1*np.random.randn(10)

x_2 = np.linspace(0,1,100)
y_2 = 2*x_2 + 0.1*np.random.randn(100)

x_3 = np.linspace(0,1,1000)
y_3 = 2*x_3 + 0.1*np.random.randn(1000)

fig = plt.figure(figsize=(5,12))
ax_1= fig.add_subplot(311)
ax_2 = fig.add_subplot(312)
ax_3 = fig.add_subplot(313)

for ax_n,x_n,y_n in zip([ax_1,ax_2,ax_3],[x_1,x_2,x_3],[y_1,y_2,y_3]):
    ax_n.plot(x_n,y_n)
    
plt.show()
```

```python
df = pd.DataFrame(data=np.random.rand(30,2),columns=['C1','C2'])
df.head(10)
```

```python
plt.figure(figsize=(12,8))
for Cn in ['C1','C2']:
    plt.plot(df[Cn])

plt.title('test',fontsize=20)
plt.xlabel('x',fontsize=15)
plt.ylabel('y',fontsize=15)

plt.legend(['data_1','data_2'])

plt.show()
```
